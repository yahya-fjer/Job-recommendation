```python
#function 
import re

# Liste des villes marocaines
moroccan_cities = [
    'Casablanca', 'Rabat', 'Fes', 'Marrakech', 'Agadir', 'Tangier', 
    'Tetouan', 'Oujda', 'Kenitra', 'Safi', 'El Jadida', 'Beni Mellal', 
    'Khouribga', 'Nador', 'Laayoune', 'Mohammedia', 'Settat', 'Salé', 
    'Ksar El Kebir', 'Taroudant','Maroc','Morroco'
    # Ajoute d'autres villes si nécessaire
]


def clean(text):
    if isinstance(text, str):
        # Supprimer les caractères spéciaux et conserver seulement les lettres, chiffres et espaces
        clean_text = re.sub(r'[^\w\s]', ' ', text)  # Retire tous les caractères sauf lettres, chiffres et espaces
        
        # Supprimer les mots constitués d'une seule lettre
        clean_text = re.sub(r'\b[a-zA-Z]\b', '', clean_text)
        
        # Supprimer les noms de villes marocaines
        for city in moroccan_cities:
            clean_text = re.sub(rf'\b{city}\b', '', clean_text, flags=re.IGNORECASE)
        
        # Supprimer les nombres
        clean_text = re.sub(r'\b\d+\b', '', clean_text)
        
        # Remplacer plusieurs espaces consécutifs par un seul espace
        clean_text = re.sub(r'\s+', ' ', clean_text)
        
        return clean_text.strip()  # Supprime les espaces en début et fin de chaîne
    else:
        return text




```

```python
sample_df['Title'] = sample_df['Title'].astype(str).apply(lambda x: clean(x))
sample_df['Description'] = sample_df['Description'].astype(str).apply(lambda x: clean(x))
```

```python
sample_df['newcollumn']=sample_df['Title'].astype(str) + " " + sample_df['Description'].astype(str)
```

```python
sample_df.reset_index(drop=True, inplace=True)
df = sample_df
```

```python
df
```

```python
from sklearn.feature_extraction.text  import  TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer()
matrix = tfidf.fit_transform(sample_df['newcollumn'])
similarity  = cosine_similarity(matrix)
```

```python
similarity
```

```python
#recommendation 
def recommandation(title):
    idx = sample_df[sample_df['Title'] == title].index[0]
    idx = sample_df.index.get_loc(idx)
    distance = sorted(list(enumerate(similarity[idx])), key=lambda x: x[1], reverse=True)[1:11]

    jobs = set()  # Utilisation d'un ensemble pour éviter les doublons
    for i in distance:
        jobs.add(df.iloc[i[0]].Title)  # Ajout au set

    return list(jobs)  # Conversion du set en liste pour retourner les résultats
```

```python
recommandation('Mobile iOS Engineer')
```
